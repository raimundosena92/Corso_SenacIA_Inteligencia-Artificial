## 1. O que são medidas de tendência central?

Medidas de tendência central são valores que tentam representar o "centro" de um conjunto de dados.

Elas ajudam a resumir informações.

Exemplos:
- qual é a nota típica de uma turma?
- qual é o salário típico de um grupo?
- qual tamanho de camiseta aparece com mais frequência?
- qual foi, em média, a quantidade vendida por dia?

As principais medidas de tendência central são:
- **média**
- **mediana**
- **moda**

Nesta aula, também veremos a **média ponderada**, que é muito importante em contextos escolares, indicadores e análise de dados.

Outras médias: Geométrica, Harmônica, Trunca, Ponto Médio e Winsorized.

## 2. Média aritmética simples

A média aritmética simples é a soma de todos os valores dividida pela quantidade de valores.

### Fórmula

$$
\bar{x} = \frac{x_1 + x_2 + x_3 + \dots + x_n}{n}
$$

### Ideia intuitiva
A média tenta "equilibrar" os dados.



### Exemplo 1

$$
\text{média} = \frac{6 + 8 + 10}{3} = \frac{24}{3} = 8

## 3. Mediana

A mediana é o valor que fica no meio dos dados **depois que eles são colocados em ordem**.



### Passo a passo
1. Ordenar os dados
2. Verificar a quantidade de elementos
3. Encontrar o valor central

### Caso 1: quantidade ímpar de dados
Exemplo: 3, 5, 8, 10, 12

O valor do meio é 8.  
Logo, a mediana é 8.

### Caso 2: quantidade par de dados
Exemplo: 4, 6, 8, 10

Os dois valores centrais são 6 e 8.  (6+8)/2=7
A mediana será a média entre eles: a medidana=7

$$
\text{mediana} = \frac{6+8}{2}=7
$$

### Fórmula para encontrar a média quando o número de elementos (n) é impar:


$$Md ={\left(\frac{n+1}{2}\right)}$$

### Fórmula para encontrar a média quando o número de elementos (n) é par:

$$Md = \frac{{\left(\frac{n}{2}\right)} + {\left(\frac{n}{2} + 1\right)}}{2}$$


### Exercício:

Crie um código python para calcular a mediana (sem utilizar bibliotecas externas, como numpy, stati, etc)

dica: para ordenar os dados, utilize ``dados ordenados = sorted(dados)``

Utilizando o numpy e statistics:

```python
# Usando IA
def mediana(lista):
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)

    if n % 2 == 0:  # n é par
        meio1 = lista_ordenada[n//2 - 1]
        meio2 = lista_ordenada[n//2]
        return (meio1 + meio2) / 2
    else:  # n é ímpar
        return lista_ordenada[n//2]

# Exemplo
dados = [2, 4, 6, 8, 10, 12]
print(mediana(dados))  # saída: 5.0

```

## 4. Moda

A moda é o valor que mais se repete.

### Exemplos

- Dados: 2, 3, 3, 5, 8  
  Moda = 3

- Dados: azul, azul, verde, vermelho, azul  
  Moda = azul

A moda é a única das medidas de tendência central que pode ser aplicada facilmente também em dados qualitativos.

### Observações importantes
- um conjunto pode ter **uma moda**: unimodal
- pode ter **duas modas**: bimodal
- pode ter **mais de duas modas**: multimodal
- pode **não ter moda**, se nenhum valor repetir

## 5. Média ponderada

Na média simples, todos os valores têm o mesmo peso.

Na média ponderada, alguns valores valem mais do que outros.

### Fórmula

$$
\text{média ponderada} = \frac{x_1p_1 + x_2p_2 + x_3p_3 + \dots + x_np_n}{p_1+p_2+p_3+\dots+p_n}
$$

onde:
- \(x\) são os valores
- \(p\) são os pesos

### Exemplo 1: notas com pesos diferentes
Um aluno tirou:
- Prova 1 = 7, peso 2
- Prova 2 = 9, peso 3

$$
\text{média ponderada} = \frac{7 \cdot 2 + 9 \cdot 3}{2+3}
= \frac{14+27}{5}
= \frac{41}{5}
= 8,2
$$

Logo, a média ponderada é 8,2.

```python
notas = [7, 9]
pesos = [2, 3]

media_ponderada = sum(n*p for n, p in zip(notas, pesos)) / sum(pesos)
media_ponderada
# zip
# n= notas
# p=peso
```

```python
import numpy as np

notas = [7, 9]
pesos = [2, 3]

media_ponderada = np.average(notas, weights=pesos)
print(media_ponderada)
```

```python
import statistics

# Cria uma lista onde o 7 aparece 2 vezes e o 9 aparece 3 vezes
dados_expandidos = [7, 7, 9, 9, 9]
media = statistics.mean(dados_expandidos)
media
#Aqui tem uma observação importante
```

### Exemplo 2: média ponderada em compras
Um produto foi vendido assim:
- 10 unidades a R\$ 5
- 30 unidades a R\$ 8

O preço médio ponderado por quantidade é:

$$
\frac{10 \cdot 5 + 30 \cdot 8}{10+30}
=
\frac{50+240}{40}
=
\frac{290}{40}
=
7,25
$$

O valor 7,25 representa melhor o preço médio real das vendas do que uma média simples entre 5 e 8.

```python
precos = [5, 8]
quantidades = [10, 30]

sum(p*q for p, q in zip(precos, quantidades)) / sum(quantidades)
```

## 6. Comparando as medidas

| Medida | Ideia principal | Ponto forte | Limitação |
|---|---|---|---|
| Média | equilíbrio dos dados | usa todos os valores | sensível a extremos |
| Mediana | valor central | resiste a outliers | não usa a intensidade completa dos dados |
| Moda | valor mais frequente | útil em categorias | pode não existir ou haver várias |
| Média ponderada | média com pesos diferentes | mais realista em muitos contextos | exige pesos corretos |

### Regra prática
- use **média** quando os dados forem relativamente equilibrados
- use **mediana** quando houver valores extremos
- use **moda** para descobrir o valor mais comum
- use **média ponderada** quando os valores tiverem importâncias diferentes

## 7. Efeito dos valores extremos

Vamos comparar dois conjuntos:

### Conjunto A
10, 12, 13, 14, 16

### Conjunto B
10, 12, 13, 14, 100

Ambos têm vários valores parecidos, mas o segundo tem um valor extremo.

```python
conjunto_a = [10, 12, 13, 14, 16]
conjunto_b = [10, 12, 13, 14, 100]

resultado = {
    "media_A": sum(conjunto_a)/len(conjunto_a),
    "mediana_A": np.median(conjunto_a),
    "media_B": sum(conjunto_b)/len(conjunto_b),
    "mediana_B": np.median(conjunto_b),
}
resultado
```

### Conclusão
No conjunto B, a média sobe bastante por causa do valor 100, mas a mediana permanece mais estável.

Isso mostra por que a mediana é tão importante em dados com outliers.

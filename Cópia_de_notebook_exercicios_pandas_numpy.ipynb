## 7. Tratamento de texto e padronização

### Exercício 13
Padronize a coluna `bolsista` para ficar apenas com:
- `Sim`
- `Não`

Dica: existem valores como `SIM`, `sim`, `Nao`.

### Exercício 14
Padronize a coluna `uf` para letras maiúsculas.

### Exercício 15
Padronize a coluna `cidade` no formato título.
Exemplo:
- `brasília` → `Brasília`


```python
# Exercício 13 - Padronize a coluna bolsista para ficar apenas com: sim, não
df['bolsista'] = df['bolsista'].str.lower().replace({
    'sim': 'Sim',
    'nao': 'Não',
    'não': 'Não'
})
df.head()
```

```python
# Exercício 14 - Padronize a coluna UF para letras maiusculas
df['uf'] = df['uf'].str.upper()
df.head()
```

```python
# Exercício 15 - Padronize a coluna cidade no formato de título
df['cidade'] = df['cidade'].str.title()
df.head()
```

## 8. Criando novas colunas

### Exercício 16
Crie uma coluna chamada `media` com a média entre `nota1` e `nota2`.

### Exercício 17
Crie uma coluna chamada `aprovado`:
- `True` para média maior ou igual a **7**
- `False` caso contrário

### Exercício 18
Crie uma coluna chamada `situacao_faltas`:
- `Alto risco` para mais de 10 faltas
- `Atenção` para 6 a 10 faltas
- `Ok` para até 5 faltas


```python
# Exercício 16 - crie uma coluna chamada media com a media da nota1 e nota2
df['media'] = (df['nota1'] + df['nota2']) / 2
df.head()
```

```python
# Exercício 17 - Crie uma coluna chamada aprovado
df['aprovado'] = df['media'] >= 7
df.head()
```

```python
# Exercício 18 - Crie uma coluna chamada situação_faltas
df['situacao_faltas'] = df['faltas'].apply(lambda x:
    'Alto risco' if x > 10 else
    'Atenção' if x >= 6 else
    'Ok'
)
df.head()
```

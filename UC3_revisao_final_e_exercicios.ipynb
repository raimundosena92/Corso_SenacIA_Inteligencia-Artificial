## 1. Função Linear

### Forma geral

$$
f(x) = ax + b
$$

### Ideia principal
A função linear representa uma relação de crescimento ou decrescimento constante.

- `a` = taxa de variação
- `b` = valor inicial

### Exemplo
$$
f(x) = 2x + 3
$$

Se `x = 4`:

$$
f(4) = 2 \cdot 4 + 3 = 11
$$

### Onde isso aparece em IA?
- regressão linear;
- previsão de valores;
- interpretação de relação entre entrada e saída.

interpretacao de relacao entre entrada e saida

```python
# Exemplo de função linear
def f_linear(x):
  return 2*x + 3
```

```python
f_linear(4)
```

```python
valores = [1,2,3,4]
```

```python
for x in valores:
  print(f'f({x}) = {f_linear(x)}')
```

```python
for x in valores:
  print(f_linear(x))
```

## 2. Função Polinomial

### Exemplo mais importante na UC3
Função do 2º grau:

$$
f(x) = ax^2 + bx + c
$$

### Ideia principal
A função polinomial permite representar comportamentos curvos.

### Exemplo
$$
f(x) = x^2 - 4x + 5
$$

### Observações
- se `a > 0`, a parábola abre para cima;
- se `a < 0`, a parábola abre para baixo.

### Onde isso aparece em IA?
- modelagem de relações não lineares;
- transformação de variáveis (`x²`, `x³`);
- regressão polinomial.

```python
# Exemplo de função polinomial
def f_polinomial(x):
  return x**2 - (4*x) + 5

```

```python
valores = [-1,0,1,2,3,4,5]
```

```python
for x in valores:
  print(f'f({x}) = {f_polinomial(x)}')
```

```python
for x in valores:
  print(f'Quando o valor de x for {x}, o resultado e:  f({x}) = {f_polinomial(x)}')
```

## 3. Função Exponencial

### Forma geral

$$
f(x) = a \cdot b^x
$$

### Ideia principal
Na função exponencial, o crescimento ou decrescimento acontece por multiplicação constante.

### Exemplo
$$
f(x) = 3 \cdot 2^x
$$

### Observações
- se `b > 1`, há crescimento;
- se `0 < b < 1`, há decrescimento.

### Onde isso aparece em IA?
- função sigmoid;
- probabilidades;
- crescimento de combinações;
- comportamento de certos algoritmos.

```python
# Exemplo de função exponencial
def f_exponencial(x):
  return 3 * (2**x)
```

```python
for x in valores:
  print(f'f({x}) = {f_exponencial(x)}')
```

## 4. Função Logarítmica

### Forma geral

$$
f(x) = \log_b(x)
$$

### Ideia principal
O logaritmo responde: “a que potência a base deve ser elevada para obter esse valor?”

### Exemplo
$$
\log_2(8) = 3
$$

porque:

$$
2^3 = 8
$$

### Onde isso aparece em IA?
- entropia;
- funções de perda;
- análise de escalas;
- complexidade de algoritmos.

```python
valores = [1,2,4,8,16]
```

```python
# Exemplo de logaritmo

import numpy as np
for x in valores:
  print(f'log_2({x}) = {np.log2(x)}')

```

```python
import math
```

```python
math.log(x, 2)
```

```python
for x in valores:
  print(f'log_2({x}) = {math.log(x, 2)}')
```

## 5. Probabilidade Básica

### Fórmula

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

### Ideias principais
- probabilidade mede chance;
- eventos independentes: multiplicamos probabilidades;
- “e” é diferente de “ou”.

### Exemplos
- probabilidade de cara em uma moeda: \(1/2\)
- probabilidade de cara duas vezes seguidas: \((1/2) \cdot (1/2) = 1/4\)

### Onde isso aparece em IA?
- classificação;
- decisão baseada em probabilidades;
- Naive Bayes;
- previsão com incerteza.

```python
# Exemplo simples de decisão da IA com probabilidades

classes = ["Cachorro", "Gato", "Onca"]
probabilidades = [0.2, 0.6, 0.2]

maior_prob = probabilidades[0]
indice_maior = 0

for i in range(len(probabilidades)):
    if probabilidades[i] > maior_prob:
        maior_prob = probabilidades[i]
        indice_maior = i

decisao = classes[indice_maior]

print("Probabilidades:", probabilidades)
print("Decisão da IA:", decisao)
```
